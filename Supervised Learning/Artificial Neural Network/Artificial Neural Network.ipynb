```python
# Importing All Important Library for ANN Modelling and Data Preprocessing 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Used to automatically split the data into tranning and test set 
from sklearn.model_selection import train_test_split

# Encoding categorical data
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Feature Scaling
from sklearn.preprocessing import StandardScaler

# This is for creating a ANN model using keras library
# Sequential Model is used to initilize our NN and Dense for creating nodes in NN
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.callbacks import EarlyStopping

# For creating confusion Matrix
from sklearn.metrics import confusion_matrix
```

```python
# Importing Dataset for Churn Analysis
dataset = pd.read_csv('Churn_Modelling.csv')
dataset.head()

# Here X and y both will be of type numpy.ndarray
X = dataset.iloc[:,3:13].values
y = dataset.iloc[:, 13].values

# Now we have a columms named as country and gender, in that we have 3 country names and two gender so for our analysis 
# we want to convert these country names into numbers such as 0, 1, 2 and 0, 1 for gender so this is done using 
# LabelEncoder function of sklearn.preprocessing 
# So 1st we need to create LabelEncoder object 

Label_Encoder_X_1 = LabelEncoder()
X[:,1] = Label_Encoder_X_1.fit_transform(X[:,1])
Label_Encoder_X_2 = LabelEncoder()
X[:,2] = Label_Encoder_X_2.fit_transform(X[:,2])
print(X)
# So after this step we are done with label encoding now we are ready for other processing
# Now as we see that the our country variable has three category as 0,1,2
# We want to split them into different columns like to_categorical in keras

OnehotEncoder = OneHotEncoder(categorical_features = [1])
X = OnehotEncoder.fit_transform(X).toarray()
print(X)
X = X[:, 1:]

# Split Data into trainning and test data and
# random_state is Pseudo-random number generator state used for random sampling
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Feature Scaling
# Creating object for feature Scaling Class
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)

# Now we are done with Data Pre-Processing now we will actually build neural network
```

```python
model = Sequential()
# Here Kernel_initializer='uniform' means when we want our model to start with appropriate random weights
# here we have input layer with dimensions=11 and then 2 hidden layer with 6 nodes each
# we are using rectified linear unit activation function which is the best in the market right now
model.add(Dense(6, kernel_initializer='uniform', activation = 'relu', input_dim= 11))
# Here we are adding dropout 10% and do not go for higher value of dropout 
# 10% of neurons will get disabled
model.add(Dropout(0.1))
model.add(Dense(6, kernel_initializer='uniform', activation='relu'))
model.add(Dropout(0.1))
# the last layer is the output layer with one node only kernal_initializer and init are same thing but use kernel_initializer
# for now onwards
# We are usign Sigmoid activation function in the last layer because we want binary outcome
# When you are dealing with more categories for classfication then use softmax function
model.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
# Compiling our ANN
# In this we have to specify the optimizer and loss function and we can also display accuracy of our model.
# Here we use adam because it it one of the best optimzer in the market because it is capable of selecting it own learning
# rate it is just like SGD but with capability mentioned above
# Now we are using binary_crossentropy loss function because we have binary classifaction problem here but we must use
# categorical_crossentropy when dealing with multiple classes in the neural network
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Now we are ready to make fit our model to dataset and make predictions
# Here the batch_size denotes by after how many examples our weights get updated
# epochs and nb_epoch are same
model.fit(X_train, y_train, batch_size=10, epochs=100)
```

```python
# Now we are making predications using our test data and model
# The predict method returns the probabilities of classification
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)
print(y_pred)
score = model.evaluate(X_test, y_test)
score
```

```python
# Confusion Matrix is something like precesion recall and true pos true neg false pos and false neg
# Now correct prediction are 1547+133 and wrong pred are 48+272 we get the same result from the evaluate method also
cm = confusion_matrix(y_test, y_pred)
cm
```

**EVALUATINING THE ARTIFICIAL NEURAL NETWORK**


```python
# Here we will use K-Fold Cross Validation in order to get a better accuracy and look about the variance
# But the problem here is that we have used Keras to build our ANN model but the K-Fold Cross Validation
# function is in scikit Learn library
# In order to incorporate scikit learn function in our keras Model we have use Keras Wrapper class to do this
# So here It is
from keras.wrappers.scikit_learn import KerasClassifier

# This is used to evaluate the score using cross_validation
from sklearn.model_selection import cross_val_score
```

```python
def build_classifier(optimizer):
    classifier = Sequential()
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
```

```python
# The only difference in the ANN build above and here is in the compilation and the training part of our model
model_1 = KerasClassifier(build_fn = build_classifier, batch_size = 10, epochs = 100)

# Till now we have not applied K-Folds Cross Validation But here we will the above step is for just 
# making the model where we can apply scikit learn K-Folds Cross Validation
# Here cv = no of K-folds in Cross Validation recommended is 10
# Here n_jobs = no of the cpu -1 is all CPU
# This will take time because it will be running 10 times
accuracies = cross_val_score(estimator=model_1, X=X_train, y=y_train, cv=10, n_jobs=-1)
```

```python
mean = accuracies.mean()
variance = accuracies.std()
print('Mean = ', mean, '\nVariance = ', variance)
```

```python
# Now here we will be using Dropout for reducing overfitting of our hypotheisis function so that it can
# generalize more towards new data and make better predictions
# You can see that your hypothesis function is overfitting the dataset when you have more accuracy over the training set
# and less accuracy on the test set
# You can also look for overfitting when you will get high variance 
# So to solve this problem of overfitting we will use Dropout Regularization
# Dropout is used above
```

**TUNING THE ANN MODEL**

```python
# We will use the technique of (Grid Search) to find a best hyperparameter for our model from a grid of values of
# hyperparameters.
# We need to first import GridSearchCV from scikit learn
from sklearn.model_selection import GridSearchCV
model_2 = KerasClassifier(build_fn=build_classifier)

# Now we will make a grid of parameters we have to make a dictionary
parameters ={'batch_size' : [25,32],
             'epochs' : [100,500],
             'optimizer': ['adam', 'rmsprop']}

# The rmsprop optimizer is used for RNN model we will study it later
grid_search = GridSearchCV(estimator=model_2, param_grid=parameters, cv=10, n_jobs=-1, scoring='accuracy')
```

Here above we have used a trick to select different optimizer you can see that in the build function above we have given
a parameter optimizer because when training the grid search wil be choosing from a list of different optimizer we have 
provided.

```python
grid_search = grid_search.fit(X_train,y_train)
# the grid_search.fit() will create a new object
best_param = grid_search.best_params_
best_accuracy = grid_search.best_score_
# We do all this inorder to acheive better accuracy
```
